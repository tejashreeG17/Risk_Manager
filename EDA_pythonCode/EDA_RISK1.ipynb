```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("risk_records1_1.csv")
```

```python
# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())
```

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe())
```

```python
# Count of records by 'Risk Type'
risk_type_counts = df['Risk Type'].value_counts()
```

```python
# Plotting the count of records by 'Risk Type'
plt.figure(figsize=(8, 6))
risk_type_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Records by Risk Type')
plt.xlabel('Risk Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plotting distribution of 'Risk Type'
plt.figure(figsize=(10, 6))
sns.countplot(x='Risk Type', data=df, palette='Set2')
plt.title('Distribution of Risk Types')
plt.xlabel('Risk Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plotting distribution of 'Industry'
plt.figure(figsize=(10, 6))
sns.countplot(x='Industry', data=df, palette='Set1')
plt.title('Distribution of Industries')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plotting distribution of 'Risk Rating'
plt.figure(figsize=(8, 6))
sns.countplot(x='Risk Rating', data=df, palette='Pastel1')
plt.title('Distribution of Risk Ratings')
plt.xlabel('Risk Rating')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
