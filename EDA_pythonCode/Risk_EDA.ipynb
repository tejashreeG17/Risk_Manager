```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("risk_records.csv")
```

```python
# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())
```

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe())
```

```python
# Count of records by 'Risk Type'
risk_type_counts = df['Risk Type'].value_counts()
```

```python
# Plotting the count of records by 'Risk Type'
plt.figure(figsize=(8, 6))
risk_type_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Records by Risk Type')
plt.xlabel('Risk Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
industry_counts = df['Industry'].value_counts()
print("\nCount of Records by Industry:")
print(industry_counts)
```

```python
risk_rating_counts = df['Risk Rating'].value_counts()
print("\nCount of Records by Risk Rating:")
print(risk_rating_counts)
```

```python
mitigation_strategy_counts = df['Mitigation Strategy'].value_counts()
print("\nDistribution of Mitigation Strategies:")
print(mitigation_strategy_counts)
```
